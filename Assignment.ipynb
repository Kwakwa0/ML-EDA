````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("GenZ_DatingApp_Data.csv")

# Task 1: Data Quality Assessment

## 1. Check for duplicate rows
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

if duplicates > 0:
    data = data.drop_duplicates()
    print("Duplicates removed.")

## 2. Check for inconsistencies in categorical values
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"Unique values in {col}: {data[col].unique()}")
    data[col] = data[col].str.strip().str.lower()

## 3. Handle missing values
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)

threshold = 0.5  # Define a threshold (50% missing values)
drop_columns = missing_values[missing_values / len(data) > threshold].index

data = data.drop(columns=drop_columns)
print(f"Dropped columns with more than {threshold*100}% missing values: {list(drop_columns)}")

data = data.fillna(data.median(numeric_only=True))
print("Filled missing values with median.")

## 4. Detect and handle outliers using the IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

numeric_columns = data.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
    outliers = detect_outliers(data, col)
    print(f"Outliers in {col}: {len(outliers)}")
    
    # Visualizing Outliers
    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[col])
    plt.title(f"Boxplot for {col}")
    plt.show()
    
    # Handling Outliers (Optional: Remove or Cap Values)
    data[col] = np.where(data[col] > outliers[col].quantile(0.95), outliers[col].quantile(0.95), data[col])
    data[col] = np.where(data[col] < outliers[col].quantile(0.05), outliers[col].quantile(0.05), data[col])
    print(f"Outliers in {col} handled.")

# Save cleaned dataset
data.to_csv("GenZ_DatingApp_Data_Cleaned.csv", index=False)
print("Cleaned dataset saved.")
````

````python
def convert_to_minutes(daily_usage_time):
    hours = 0
    minutes = 0

    hours_match = re.search(r"(\d+|\d+\.\d+)\s*hour", daily_usage_time)
    if hours_match:
        hours = float(hours_match.group(1))

    minutes_match = re.search(r"(\d+)\*minute", daily_usage_time)
    if minutes_match:
        minutes = int(minutes_match.group(1))

    return int((hours*60) + minutes)
````
